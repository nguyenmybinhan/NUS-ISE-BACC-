```python
import numpy as np
import pandas as pd
```

```python
# Load Full Demand data
demand_df = pd.read_csv(r"P_VaaT_Supporting_Document_Q2.2_Full_Demand.csv")
demand_df["Demand"] = demand_df["Demand"].astype(int)
demand_df.head()
```

```python
# Load Machine Capacity data
machine_capacity_df = pd.read_csv(r"P_VaaT_Supporting_Document_Q2.2._Machines_Capacity.csv")
machine_capacity_df["MachineCapacity"] = machine_capacity_df["MachineCapacity"].astype(int)
machine_capacity_df.head()
```

```python
# Initializing answer DataFrame
machine_calc_df = pd.DataFrame(columns=["Workstation"])
machine_calc_df["Workstation"] = machine_capacity_df["Workstation"]

# Calculate the number of machines for each demand (each row)
for idx in range(demand_df.shape[0]):
    col_name = demand_df["Workweek"][idx]
    machine_calc_df[col_name] = np.ceil(demand_df["Demand"][idx]/machine_capacity_df["MachineCapacity"])
    machine_calc_df[col_name] = machine_calc_df[col_name].astype(int)

machine_calc_df.head()
```

```python
# Export DataFrame to a csv file
machine_calc_df.to_csv("P_VaaT_Supporting_Document_Q2.2_Machines_Calc.csv", index=False)
```
